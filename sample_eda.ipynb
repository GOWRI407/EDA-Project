##CHARTS

```python
print("\n🔹 Correlation Heatmap:")# to display values in heatmap
numerical_df = df.select_dtypes(include=[np.number])
plt.figure(figsize=(12, 10)) # Adjust figure size for better clarity
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation between Numerical Features")
plt.show()
```

##PRICE DISTRIBUTION

```python
sns.histplot(df['price'], bins=30, kde=True)
plt.title('Distribution of Smartphone Prices')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

##Smartphone Prices by Brand

```python
sns.boxplot(x='brand_name', y='price', data=df)
plt.xticks(rotation=45)
plt.title('Smartphone Prices by Brand')
plt.show()
```

##Battery Capacity vs Price

```python
sns.scatterplot(x='battery_capacity', y='price', data=df)
plt.title('Battery Capacity vs Price')
plt.xlabel('Battery Capacity (mAh)')
plt.ylabel('Price')
plt.show()
```

##Top 10 Brands by Number of Smartphones

```python
df['brand_name'].value_counts().nlargest(10).plot(kind='bar')
plt.title('Top 10 Brands by Number of Smartphones')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

##RAM vs Price by Brand

```python
sns.scatterplot(data=df[df['price'] <= 150000], x='ram_capacity', y='price', hue='brand_name', palette='tab20', s=70)
plt.title("RAM vs Price by Brand")
plt.legend(bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

##Internal Memory Distribution

```python
sns.countplot(x='internal_memory', data=df)
plt.title('Internal Memory Distribution')
plt.xlabel('Internal Memory (GB)')
plt.ylabel('Number of Smartphones')
plt.xticks(rotation=45)
plt.show()
```

##Price vs Screen Size

```python
sns.scatterplot(x='screen_size', y='price', data=df)
plt.title('Price vs Screen Size')
plt.xlabel('Screen Size (inches)')
plt.ylabel('Price')
plt.show()
```

##Number of Smartphones per RAM Size

```python
sns.countplot(x='ram_capacity', data=df, order=sorted(df['ram_capacity'].unique()))
plt.title('Number of Smartphones per RAM Size')
plt.xlabel('RAM (GB)')
plt.ylabel('Count')
plt.show()
```

```python
sns.pairplot(df[['price', 'battery_capacity', 'ram_capacity', 'internal_memory']], diag_kind='kde')
plt.suptitle("Pairwise Relationships", y=1.02)
plt.show()
```

##Price Distribution by OS


```python
g = sns.FacetGrid(df, col='os', col_wrap=3, height=4)
g.map(sns.histplot, 'price', bins=20, kde=True)
plt.suptitle("Price Distribution by OS", y=1.05)
plt.show()
```

##Price vs RAM grouped by OS

```python
sns.swarmplot(x='ram_capacity', y='price', hue='os', data=df)
plt.title("Price vs RAM grouped by OS")
plt.legend(bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

```python
import squarify
counts = df['brand_name'].value_counts()
plt.figure(figsize=(12, 6))
squarify.plot(sizes=counts.values, label=counts.index, color=sns.color_palette('Set3', len(counts)), alpha=0.9, text_kwargs={'fontsize':10})
plt.title("Smartphone Counts by Brand (Treemap)", fontsize=14); plt.axis('off'); plt.tight_layout(); plt.show()
```

##Stacked Bar: OS Distribution by Brand

```python
pd.crosstab(df['brand_name'], df['os']).plot(kind='bar', stacked=True, colormap='viridis', figsize=(12,6))
plt.title("Stacked Bar: OS Distribution by Brand")
plt.xticks(rotation=45)
plt.ylabel("Number of Phones")
plt.tight_layout()
plt.show()
```

##Interactive: Battery vs Price by Brand

```python
import plotly.express as px

fig = px.scatter(df, x="battery_capacity", y="price", color="brand_name", size="ram_capacity", hover_data=["model"])
fig.update_layout(title="Interactive: Battery vs Price by Brand")
fig.show()
```

##Radar: Avg Specs by Brand

```python
avg = df[df['brand_name'].isin(df['brand_name'].value_counts().nlargest(5).index)].groupby('brand_name')[['ram_capacity','battery_capacity','internal_memory']].mean()/df[['ram_capacity','battery_capacity','internal_memory']].max()

angles = np.linspace(0, 2*np.pi, 3, endpoint=False).tolist()
angles += angles[:1]

plt.figure(figsize=(7,7))
for b in avg.index:
    vals = avg.loc[b].tolist()
    vals += vals[:1]
    plt.polar(angles, vals, label=b)
    plt.fill(angles, vals, alpha=0.1)

plt.xticks(angles[:-1], ['RAM','Battery','Storage'])
plt.title("Radar: Avg Specs by Brand")
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python



fig, axes = plt.subplots(2, 3, figsize=(22, 12))
sns.set(style="whitegrid")

sns.scatterplot(data=df[df['price'] <= 150000], x='ram_capacity', y='price',
                hue='brand_name', palette='tab20', s=60, ax=axes[0, 0])
axes[0, 0].set_title("RAM vs Price by Brand", fontsize=14)
axes[0, 0].legend(loc='best', fontsize=8)


sns.swarmplot(x='ram_capacity', y='price', hue='os', data=df, ax=axes[0, 1])
axes[0, 1].set_title("Price vs RAM grouped by OS", fontsize=14)
axes[0, 1].legend(loc='best', fontsize=8)


counts = df['brand_name'].value_counts()
squarify.plot(sizes=counts.values,
              label=counts.index,
              color=sns.color_palette('Set3', len(counts)),
              alpha=0.9,
              text_kwargs={'fontsize': 9},
              ax=axes[0, 2])
axes[0, 2].set_title("Smartphone Counts by Brand (Treemap)")
axes[0, 2].axis('off')


pd.crosstab(df['brand_name'], df['os']).plot(kind='bar', stacked=True,
                                             colormap='viridis', ax=axes[1, 0])
axes[1, 0].set_title("Stacked Bar: OS Distribution by Brand")
axes[1, 0].set_ylabel("Number of Phones")
axes[1, 0].tick_params(axis='x', rotation=45)

avg = df[df['brand_name'].isin(df['brand_name'].value_counts().nlargest(5).index)] \
        .groupby('brand_name')[['ram_capacity','battery_capacity','internal_memory']] \
        .mean() / df[['ram_capacity','battery_capacity','internal_memory']].max()

angles = np.linspace(0, 2 * np.pi, len(avg.columns), endpoint=False).tolist()
angles += angles[:1]

ax_radar = plt.subplot(2, 3, 5, polar=True)
for brand in avg.index:
    values = avg.loc[brand].tolist()
    values += values[:1]
    ax_radar.plot(angles, values, label=brand)
    ax_radar.fill(angles, values, alpha=0.1)
ax_radar.set_xticks(angles[:-1])
ax_radar.set_xticklabels(['RAM', 'Battery', 'Storage'])
ax_radar.set_title("Radar: Avg Specs by Top Brands", y=1.1)
ax_radar.legend(loc='upper right', bbox_to_anchor=(1.4, 1))


axes[1, 2].axis('off')
axes[1, 2].text(0.5, 0.5, 'Plotly Interactive Chart Below ⬇️',
               horizontalalignment='center', verticalalignment='center', fontsize=12)

plt.tight_layout()
plt.show()

fig = px.scatter(df, x="battery_capacity", y="price", color="brand_name",
                 size="ram_capacity", hover_data=["model"],
                 title="Interactive: Battery vs Price by Brand")
fig.show()
```
